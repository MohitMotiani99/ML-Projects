```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('neph_data.csv')
```

```python
data
```

```python
data=data.fillna('None')
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder=LabelEncoder()
data=label_encoder.fit_transform(data)
```

```python
data['Nausea']=label_encoder.fit_transform(data['Nausea'])
```

```python
data
```

```python
data['Lumber Pain']=label_encoder.fit_transform(data['Lumber Pain'])
data['Urine Pushing']=label_encoder.fit_transform(data['Urine Pushing'])
data['Micturartion Pains']=label_encoder.fit_transform(data['Micturartion Pains'])
data['Burning Urethra']=label_encoder.fit_transform(data['Burning Urethra'])
data['Urinary Bladder Inflammation']=label_encoder.fit_transform(data['Urinary Bladder Inflammation'])
data['Nephritis']=label_encoder.fit_transform(data['Nephritis'])
```

```python
data
```

```python
datax=data[['Temp','Nausea','Lumber Pain','Urine Pushing','Micturartion Pains','Burning Urethra']].values
```

```python
datay=data['Nephritis']
```

```python
datax
```

```python
datay
```

```python
from sklearn.model_selection import train_test_split

trainx,testx,trainy,testy=train_test_split(datax,datay,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier(criterion='entropy',max_depth=10)
model.fit(trainx,trainy)
```

```python
model
```

```python
predy=model.predict(testx)
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy is: ",accuracy_score(testy,predy)*100)
```
