```python
for i in range(train.shape[0]):
    if train.loc[i,'Item_Fat_Content'][0]=='l' or train.loc[i,'Item_Fat_Content'][0]=='L' :
        train.loc[i,'Item_Fat_Content']='Low Fat'
    elif train.loc[i,'Item_Fat_Content'][0]=='r'or train.loc[i,'Item_Fat_Content'][0]=='R' :
        train.loc[i,'Item_Fat_Content']='Regular'
```

```python
from sklearn.preprocessing import LabelEncoder

order_fat_content={}
order_fat_content['None']=0

label_encoder=LabelEncoder()

train['Item_Fat_Content']=train['Item_Fat_Content'].fillna('None')
unique_fat_content_values=train['Item_Fat_Content'].unique()
print(train['Item_Fat_Content'].unique())
for i in range(len(unique_fat_content_values)):
    order_fat_content[unique_fat_content_values[i]]=i+1
    

print(order_fat_content)


```

```python
train['Item_Fat_Content']=train['Item_Fat_Content'].map(order_fat_content)

train
```

```python
train['Item_Type']=train['Item_Type'].fillna('None')

order_item_type={}
order_item_type['None']=0

unique_item_type_values=train['Item_Type'].unique()
for i in range(len(unique_item_type_values)):
    order_item_type[unique_item_type_values[i]]=i+1;

print(unique_item_type_values)
print(order_item_type)

```

```python
train['Item_Type']=train['Item_Type'].map(order_item_type)

train
```

```python
train['Item_Weight']=train['Item_Weight'].fillna(0.0)
train['Item_Visibility']=train['Item_Visibility'].fillna(0.0)
train['Item_MRP']=train['Item_MRP'].fillna(0.0)
```

```python
train
```

```python
train.loc[7,:]
```

```python
trainx=train[['Item_Weight','Item_Fat_Content','Item_Visibility','Item_Type','Item_MRP']].values
```

```python
from sklearn.preprocessing import StandardScaler

scalar=StandardScaler()

trainx=scalar.fit_transform(trainx)
```

```python
trainx
```

```python
trainy=train['Item_Outlet_Sales'].values
```

```python
trainy=trainy.reshape(trainy.shape[0],1)

trainy=scalar.fit_transform(trainy)
```

```python
trainy
```

```python
from sklearn.model_selection import train_test_split

trainx,validx,trainy,validy=train_test_split(trainx,trainy,test_size=0.2)
```

```python
trainx[:][1]
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score

acc_scores=np.zeros((60));

for k in range(1,61):
    model=KNeighborsRegressor(n_neighbors=k)
    model.fit(trainx,trainy)
    
    predy=model.predict(validx)
    acc_scores[k-1]=r2_score(validy,predy)
    
```

```python
plt.plot(range(1,61),acc_scores,'bo')
plt.show()
```
