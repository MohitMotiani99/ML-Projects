```python
cat_features=[]
for col in cols:
    if house_data[col].isna().sum()!=0:
        if house_data[col].dtype=='object':
            house_data[col].fillna('missing',inplace=True)
            cat_features.append(col)
        elif house_data[col].dtype=='int64':
            s=house_data[col].value_counts()
            house_data[col].fillnana(s.index[0],inplace=True)
            #cat_features.append(col)
        else:
            house_data[col].fillna(house_data[col].mean(),inplace=True)
    else:
        if house_data[col].dtype=='object':
            cat_features.append(col)
            
            
```

```python
X=house_data.drop('SalePrice',axis=1)
y=house_data['SalePrice']
```

```python
#encoding

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

one_hot=OneHotEncoder(sparse=False)
transformer=ColumnTransformer([
    ('one_hot',one_hot,cat_features)
],
                             remainder='passthrough')

transformed_X=transformer.fit_transform(X)
```
