```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston=load_boston()
```

```python
boston
```

```python
boston_df=pd.DataFrame(boston['data'],columns=boston['feature_names'])
boston_df
```

```python
boston_df['Price']=pd.Series(boston['target'])
boston_df
```

```python
# checking for non numerical data
boston_df.dtypes
```

No non numerical data so feature encoding not required

```python
#check for missing values in Price col and drop if any
boston_df['Price'].isna().sum()
```

No null values in Price so go ahead

```python
#separating features and labels
X=boston_df.drop('Price',axis=1)
y=boston_df['Price']
```

```python
X.head()
```

```python
y.head()
```

```python
np.random.seed(42)

#split the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
#Choose the right model using sk map-->Ridge
np.random.seed(42)
from sklearn.linear_model import Ridge

model=Ridge()
model.fit(X_train,y_train)
#model.score(X_test,y_test)
print(f'Accuracy using Ridge Regression is {model.score(X_test,y_test)*100:.2f}% ')
```

We can improve the Accuracy so back to the map

```python
#Try ensemble method RandomForestRegressor
np.random.seed(42)

from sklearn.ensemble import RandomForestRegressor

estimators=np.arange(100,1000,50)
acc=[]
for i in estimators:
    rf=RandomForestRegressor(n_estimators=i)
    rf.fit(X_train,y_train)
    print(f'Random Forest Regressor with {i} estimators')
    print(f'Accuracy using Random Forest Regressor with {i} estimators is {rf.score(X_test,y_test)*100:.2f}%')
    acc.append(rf.score(X_test,y_test)*100)
    print()
```

```python
fig,ax=plt.subplots(figsize=(20,10))
ax.plot(estimators,acc);
ax.set(title='Accuracy of Random Forest Regressor with varying number of estimators',
      xlabel='No. of Estimators',
      );
ax.yaxis.set_ticks(np.arange(65,110,10));
ax.xaxis.set_ticks(np.arange(100,1000,50));
ax.axhline(y=model.score(X_test,y_test),linestyle='--');
```

```python
fig,ax=plt.subplots(figsize=(20,10))
ax.plot(estimators,acc);
ax.set(title='Accuracy of Random Forest Regressor with varying number of estimators',
      xlabel='No. of Estimators',
      ylabel='Accuracy',
      );

```

```python
acc
```

```python
acc_dict={
}
for k,v in zip(estimators,acc):
    acc_dict[k]=v
```

```python
acc_dict
```

```python
best_estimator=max(acc_dict,key=lambda x:acc_dict[x])
```

```python
best_estimator
```

```python
print(f'Best Random Forest Regressor is with {best_estimator} estimators')
```

```python
np.random.seed(42)
rf=RandomForestRegressor(n_estimators=best_estimator)
rf.fit(X_train,y_train)
print(f'Accuracy of best Random Forest Regressor is {rf.score(X_test,y_test)*100:.2f}%')
```
